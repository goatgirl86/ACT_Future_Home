```python
import pandas as pd
```

```python
year_list_1 = ['12', '13', '14', '15', '16', '17', '18']
year_list_2 = ['19', '20', '21']
```

```python
final_columns = ['YEAR', 'AREA', 'AREA_TITLE', 'OCC_CODE', 'OCC_TITLE', 'OCC_GROUP', 'TOT_EMP', 'EMP_PRSE', 'JOBS_1000', 'LOC_QUOTIENT', 'H_MEAN', 'A_MEAN', 'MEAN_PRSE', 'H_PCT10', 'H_PCT25', 'H_MEDIAN', 'H_PCT75', 'H_PCT90', 'A_PCT10', 'A_PCT25', 'A_MEDIAN', 'A_PCT75', 'A_PCT90', 'ANNUAL', 'HOURLY']
len(final_columns)
```

```python
for year in year_list_1:
    
    sheet = 'state_dl'

    if year == '12' or year =='13':
        file_path = 'oesm' + year + 'st/' + 'state_M20' + year + '_dl.xls'
        sheet = 'state_dl'
    else:
        file_path = 'oesm' + year + 'st/' + 'state_M20' + year + '_dl.xlsx'
    
    
    df = pd.read_excel(file_path, sheet_name=sheet)
    
    df = df.drop(columns=['ST'])
    df['Year'] = '20' + year
    df = df.loc[df['OCC_GROUP'] == 'major']
    #df = df.loc[((df['OCC_GROUP'] == 'major') & (df['STATE'] != 'Guam') & (df['STATE'] != 'Puerto Rico') & (df['STATE'] != 'Virgin Islands'))]

    if year == '12':
        occ_emp_df = df.copy()
        print(year)
    else:
        occ_emp_df = pd.concat([occ_emp_df, df])
        print(year)

occ_emp_df.columns = ['AREA', 'AREA_TITLE', 'OCC_CODE', 'OCC_TITLE', 'OCC_GROUP', 'TOT_EMP', 'EMP_PRSE', 'JOBS_1000', 'LOC_QUOTIENT', 'H_MEAN', 'A_MEAN', 'MEAN_PRSE', 'H_PCT10', 'H_PCT25', 'H_MEDIAN', 'H_PCT75', 'H_PCT90', 'A_PCT10', 'A_PCT25', 'A_MEDIAN', 'A_PCT75', 'A_PCT90', 'ANNUAL', 'HOURLY', 'YEAR']
occ_emp_df = occ_emp_df[final_columns]

# missing 2012 STATE: 'District of Columbia', OCC_TITLE': 'Farming, Fishing, and Forestry Occupations' row
```

```python
for year in year_list_2:
    file_path = 'oesm' + year + 'st/' + 'state_M20' + year + '_dl.xlsx'
    
    if year == '19':
        sheet = 'State_M20' + year + '_dl'
        df = pd.read_excel(file_path, sheet_name=sheet)
        df = df.drop(columns = ['area_type', 'naics', 'naics_title', 'i_group', 'own_code', 'pct_total'])
        df['Year'] = '20' + year
        df = df.loc[df['o_group'] == 'major']
        #df = df.loc[((df['o_group'] == 'major') & (df['area_title'] != 'Guam') & (df['area_title'] != 'Puerto Rico') & (df['area_title'] != 'Virgin Islands'))]
        occ_emp_df_2 = df.copy()
        occ_emp_df_2.columns = ['AREA', 'AREA_TITLE', 'OCC_CODE', 'OCC_TITLE', 'OCC_GROUP', 'TOT_EMP', 'EMP_PRSE', 'JOBS_1000', 'LOC_QUOTIENT', 'H_MEAN', 'A_MEAN', 'MEAN_PRSE', 'H_PCT10', 'H_PCT25', 'H_MEDIAN', 'H_PCT75', 'H_PCT90', 'A_PCT10', 'A_PCT25', 'A_MEDIAN', 'A_PCT75', 'A_PCT90', 'ANNUAL', 'HOURLY', 'YEAR']

    elif year == '20':
        sheet = 'State_M20' + year + '_dl'
        df = pd.read_excel(file_path, sheet_name=sheet)
        df = df .drop(columns = ['AREA_TYPE', 'PRIM_STATE', 'NAICS', 'NAICS_TITLE', 'I_GROUP', 'OWN_CODE', 'PCT_TOTAL'])
        df['Year'] = '20' + year
        df = df.loc[df['O_GROUP'] == 'major']
        #df = df.loc[((df['O_GROUP'] == 'major') & (df['AREA_TITLE'] != 'Guam') & (df['AREA_TITLE'] != 'Puerto Rico') & (df['AREA_TITLE'] != 'Virgin Islands'))]
        df.columns = ['AREA', 'AREA_TITLE', 'OCC_CODE', 'OCC_TITLE', 'OCC_GROUP', 'TOT_EMP', 'EMP_PRSE', 'JOBS_1000', 'LOC_QUOTIENT', 'H_MEAN', 'A_MEAN', 'MEAN_PRSE', 'H_PCT10', 'H_PCT25', 'H_MEDIAN', 'H_PCT75', 'H_PCT90', 'A_PCT10', 'A_PCT25', 'A_MEDIAN', 'A_PCT75', 'A_PCT90', 'ANNUAL', 'HOURLY', 'YEAR']
        occ_emp_df_2 = pd.concat([occ_emp_df_2, df])
    else:
        sheet = 'All May 2021 data'
        df = pd.read_excel(file_path, sheet_name=sheet)
        df = df .drop(columns = ['AREA_TYPE', 'PRIM_STATE', 'NAICS', 'NAICS_TITLE', 'I_GROUP', 'OWN_CODE', 'PCT_TOTAL','PCT_RPT'])
        df['Year'] = '20' + year
        df = df.loc[df['O_GROUP'] == 'major']
        #df = df.loc[((df['O_GROUP'] == 'major') & (df['AREA_TITLE'] != 'Guam') & (df['AREA_TITLE'] != 'Puerto Rico') & (df['AREA_TITLE'] != 'Virgin Islands'))]
        df.columns = ['AREA', 'AREA_TITLE', 'OCC_CODE', 'OCC_TITLE', 'OCC_GROUP', 'TOT_EMP', 'EMP_PRSE', 'JOBS_1000', 'LOC_QUOTIENT', 'H_MEAN', 'A_MEAN', 'MEAN_PRSE', 'H_PCT10', 'H_PCT25', 'H_MEDIAN', 'H_PCT75', 'H_PCT90', 'A_PCT10', 'A_PCT25', 'A_MEDIAN', 'A_PCT75', 'A_PCT90', 'ANNUAL', 'HOURLY', 'YEAR']
        occ_emp_df_2 = pd.concat([occ_emp_df_2, df])
    print(year)


occ_emp_df_2 = occ_emp_df_2[final_columns]

```

```python
occ_emp_df_2
```

```python
final_occ_emp_df = pd.concat([occ_emp_df, occ_emp_df_2])
final_occ_emp_df
```

```python
region_df = pd.read_csv('regions.csv')

def get_region(area_title):
    return region_df.loc[region_df['Area title']==area_title]['Region'].values[0]
```

```python
final_occ_emp_df['REGION'] = final_occ_emp_df['AREA_TITLE'].apply(lambda x: get_region(x))
final_occ_emp_df = final_occ_emp_df[['YEAR', 'REGION', 'AREA', 'AREA_TITLE', 'OCC_CODE', 'OCC_TITLE', 'OCC_GROUP', 'TOT_EMP', 'EMP_PRSE', 'JOBS_1000', 'LOC_QUOTIENT', 'H_MEAN', 'A_MEAN', 'MEAN_PRSE', 'H_PCT10', 'H_PCT25', 'H_MEDIAN', 'H_PCT75', 'H_PCT90', 'A_PCT10', 'A_PCT25', 'A_MEDIAN', 'A_PCT75', 'A_PCT90', 'ANNUAL', 'HOURLY']]
final_occ_emp_df
```

```python
final_occ_emp_df.to_csv('occupational_employment_wage(state).csv', index=False)
```
