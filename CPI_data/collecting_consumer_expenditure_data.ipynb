### collecting consumer expenditure data stored in the Original_CPI_data/consumer_expenditure, converting the data to the programmable format and storing it in consumer_expenditure_region_2020-2021.csv.
### (The data in the Original_CPI_data/consumer_expenditure folder is from https://www.bls.gov/cex/tables.htm)

```python
import pandas as pd
```

```python
file_path = 'Original_CPI_data/consumer_expenditure/region/cu-region-2-year-average-2021.xlsx'

df = pd.read_excel(file_path, sheet_name='Table 1800')
df
```

```python
column_index = df.index[df['Table 1800. Region of residence: Annual expenditure means, standard errors, and coefficients of variation, Consumer Expenditure Surveys, 2020-2021'] == 'Item'].values[0]
column_index
```

```python
column_list = df.loc[column_index,:].values.tolist()
column_list[1] = 'All'
column_list
```

```python
df = df.loc[column_index+1:,:]
df.columns = column_list
null_list = df.index[df['Item'].isnull()].values.tolist()
df = df.drop(null_list)
df.reset_index(drop=True,inplace=True)
df
```

```python
Item_list = ['Income before taxes','Income after taxes','Average annual expenditures','Food','Alcoholic beverages','Housing','Apparel and services','Transportation','Healthcare','Entertainment','Personal care products and services','Reading','Education','Tobacco products and smoking supplies','Miscellaneous','Cash contributions','Personal insurance and pensions']
item_index = []
```

```python
for item in Item_list:
    keyword_index = df.index[df['Item'] == item].values[0]
    values = df.loc[keyword_index+1,:].values.tolist()
    values[0] = item
    df.loc[df['Item'] == item] = values
    i = df.index[df['Item'] == item].values[0]
    item_index.append(i)

df
```

```python
print(len(item_index))
item_index
```

```python
new_df = df.iloc[item_index]
new_df = new_df.set_index('Item')
new_df = new_df.T
new_df.columns.name = ''
new_df.reset_index(inplace=True)
new_df.rename(columns={'index':'Region'}, inplace=True)
new_df
```

```python
new_df.to_csv('consumer_expenditure_region_2020-2021.csv', index=False)
```
