```python
import pandas as pd
```

```python
year_list = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
```

```python
for year in year_list:
    file_path = 'consumer_expenditure/region/cu-region-2-year-average-'+ str(year) + '.xlsx'
    #print (file_path)
    df = pd.read_excel(file_path, sheet_name='Table 1800')

    if year != 2019 and year != 2020 and year != 2021:
        index_column = 'Table 1800. Region of residence: Average annual expenditures and characteristics, Consumer Expenditure Survey, ' +str(year-1) + '-' + str(year)
    elif year == 2019:
        index_column = 'Table 1800. Region of residence: Annual expenditure means, standard errors, and coefficients of variation, Consumer Expenditure Survey, ' +str(year-1) + '-' + str(year)
    else:
        index_column = 'Table 1800. Region of residence: Annual expenditure means, standard errors, and coefficients of variation, Consumer Expenditure Surveys, ' + str(year-1) + '-' + str(year)
        
    column_index = df.index[df[index_column] == 'Item'].values[0]
    
    column_list = df.loc[column_index,:].values.tolist()
    column_list[1] = 'All'
    
    df = df.loc[column_index+1:,:]

    df.columns = column_list
    null_list = df.index[df['Item'].isnull()].values.tolist()
    df = df.drop(null_list)
    df.reset_index(drop=True,inplace=True)
    #print(df.shape)

    if year == 2012:
        Item_list = ['Income before taxes','Income after taxes','Average annual expenditures','Food','Alcoholic beverages','Housing','Apparel and services','Transportation','Health care','Entertainment','Personal care products and services','Reading','Education','Tobacco products and smoking supplies','Miscellaneous','Cash contributions','Personal insurance and pensions']
    else:
        Item_list = ['Income before taxes','Income after taxes','Average annual expenditures','Food','Alcoholic beverages','Housing','Apparel and services','Transportation','Healthcare','Entertainment','Personal care products and services','Reading','Education','Tobacco products and smoking supplies','Miscellaneous','Cash contributions','Personal insurance and pensions']
    
    item_index = []
    
    # get indexes of item and replace the row of item with the values of item
    for item in Item_list:
        keyword_index = df.index[df['Item'] == item].values[0]
        if year != 2019 and year != 2020 and year != 2021:
            values = df.loc[keyword_index,:].values.tolist()
        else:
            values = df.loc[keyword_index+1,:].values.tolist()
            values[0] = item
            df.loc[df['Item'] == item] = values
        """    
        values[0] = item
        df.loc[df['Item'] == item] = values
        """
        i = df.index[df['Item'] == item].values[0]
        item_index.append(i)
    
    #print(len(item_index))
    df = df.iloc[item_index]
    df = df.set_index('Item')
    df = df.T
    df.columns.name = ''
    df.reset_index(inplace=True)
    df.rename(columns={'index':'Region'}, inplace=True)
    df['Year'] = str(year)
    #print(df)

    if year == 2012:
        df.rename(columns={'Health care': 'Healthcare'}, inplace=True)
        new_df = df.copy()
    else:
        new_df = pd.concat([new_df, df])
    
```
