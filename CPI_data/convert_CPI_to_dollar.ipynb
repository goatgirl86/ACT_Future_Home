### loading CPI data (CPIs_data_2012-2022Oct.csv) and consumer expenditure data for regions (consumer_expenditure_region_2020-2021.csv), converting the CPI data to dollar amounts using consumer expenditure data and storing the data in consumer_expenditure_2012-2022(dollar).csv.

```python
import pandas as pd
```

```python
CPI_file_path = 'CPIs_data_2012-2022Oct.csv'
consumer_expenditure_file = 'consumer_expenditure_region_2020-2021.csv'
```

```python
CPI_df = pd.read_csv(CPI_file_path)
CPI_df
```

```python
expenditure_df = pd.read_csv(consumer_expenditure_file)
temp = expenditure_df.iloc[0]
expenditure_df.iloc[0] = expenditure_df.iloc[4]
expenditure_df.iloc[4] = temp
expenditure_df
```

```python
Annual_CPI_df = CPI_df.loc[(CPI_df['Month'] == 'Annual') & ((CPI_df['Year'] == 2020) | (CPI_df['Year'] == 2021))].copy()
Annual_CPI_df
```

```python
Average_CPI_df = pd.DataFrame()
region_list = ['Western','Northeast','Midwest','South','All']

Average_CPI_df['Region'] = region_list
Average_CPI_df['Year'] = '2020-2021'
Average_CPI_df['Month'] = 'average'

column_list = ['Food','Alcoholic_beverages','Housing', 'Apparel','Transportation', 'Healthcare', 'Entertainment','Education_and_communication','Other_goods_and_services']

for column in column_list:
    data = []
    for region in region_list:
        average_cpi = Annual_CPI_df.loc[Annual_CPI_df['Region'] == region][column].mean()
        data.append(average_cpi)
    Average_CPI_df[column] = data

Average_CPI_df
```

```python
expense_df = expenditure_df.drop(columns=['Income before taxes','Income after taxes','Average annual expenditures'])
expense_df['Education'] = expense_df['Education'] + expense_df['Reading']
expense_df['Other_goods_and_services'] = expense_df['Personal care products and services'] + expense_df['Tobacco products and smoking supplies'] + expense_df['Miscellaneous'] + expense_df['Cash contributions']+ expense_df['Personal insurance and pensions']
expense_df = expense_df.drop(columns=['Reading','Personal care products and services','Tobacco products and smoking supplies','Miscellaneous','Cash contributions','Personal insurance and pensions'])
expense_df = expense_df.rename(columns={'Alcoholic beverages':'Alcoholic_beverages','Apparel and services':'Apparel', 'Education':'Education_and_communication'})
expense_df
```

```python
annual_original_expense_df = Average_CPI_df.copy()
monthly_original_expense_df = Average_CPI_df.copy()
annual_original_expense_df
```

```python
for column in column_list:
    annual_original_expense_df[column] = expense_df[column] / annual_original_expense_df[column] * 100 # annual expense
    annual_original_expense_df[column] = annual_original_expense_df[column].round(2)

    monthly_original_expense_df[column] = expense_df[column] / monthly_original_expense_df[column] * 100 /12 # monthly expense
    monthly_original_expense_df[column] = monthly_original_expense_df[column].round(2)

monthly_original_expense_df
```

```python
monthly_original_expense_df['Year'] = '1982-1984'
monthly_original_expense_df['Month'] = 'original amount'
monthly_original_expense_df
```

```python
annual_original_expense_df['Year'] = '1982-1984'
annual_original_expense_df['Month'] = 'original amount'

annual_original_expense_df.to_csv('annual_original_expense.csv', index=False)
```

```python
consumer_expenditure_df = CPI_df.copy()
```

```python
for column in column_list:
    data_list = []
    for region in region_list:
        data_list.extend((CPI_df.loc[CPI_df['Region'] == region][column].mul(monthly_original_expense_df.loc[monthly_original_expense_df['Region'] == region][column].values[0])/100).round(2).tolist())
        print(region)
    print(column)
    print(len(data_list))
    consumer_expenditure_df[column] = data_list

consumer_expenditure_df
```

```python
consumer_expenditure_df.to_csv('consumer_expenditure_2012-2022(dollar).csv',index=False)
```
