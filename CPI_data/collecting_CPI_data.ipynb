### collecting CPI data stored in the Original_CPI_data, converting the data to the programmable format and storing it in CPIs_data_2012-2022Oct.csv.
### (The data in the Original_ CPI_data folder is from https://www.bls.gov/cpi/data.htm)

```python
import pandas as pd
```

```python
files_list = ['Food','Alcoholic_beverages','Housing', 'Apparel','Transportation', 'Healthcare', 'Entertainment','Education_and_communication','Other_goods_and_services']
region_list = ['Western','Northeast','Midwest','South','All']
```

```python
for region in region_list:    
    for file in files_list:
        directory = 'Original_CPI_data/' + region + '/'
        file_extension = '_2012-2022Oct.xlsx'
        path = directory + file + file_extension

        df = pd.read_excel(path, sheet_name='BLS Data Series')

        column_index = df.index[df['CPI for All Urban Consumers (CPI-U)'] == 'Year'].values[0]

        column_list = df.loc[column_index,:].values.tolist()

        df = df.loc[column_index+1:,:]
        df.columns = column_list
        df = df.drop(columns=['HALF1','HALF2'])
        df['Year'] = df['Year'].astype(int)
        df.reset_index(inplace=True, drop=True)

        df.set_index('Year',inplace=True)
        df = df.stack()
        df = df.reset_index()
        df.columns = ['Year','Month', file]

        if file == 'Food':
            cpi_df = df.copy()
        else:
            cpi_df = cpi_df.merge(df, how='left', on=['Year','Month'])

    cpi_df['Region'] = region
    if region == 'Western':
        cpi_df = cpi_df[['Region','Year','Month','Food','Alcoholic_beverages','Housing', 'Apparel','Transportation', 'Healthcare', 'Entertainment','Education_and_communication','Other_goods_and_services']]
        CPIs_df = cpi_df.copy()
    else:
        CPIs_df = pd.concat([CPIs_df, cpi_df])
```
